```python
import pandas as pd
```

```python
from datetime import datetime
from datetime import timedelta
import random
from random import seed
from random import random


df = pd.DataFrame({
    'temp': [],
    'dewpoint': [],    
    'date': []
})

t = []
print(type(t))
print(len(t))
temperature = []
dewpoint = []

sampleDate = datetime(2020,1, 2, 0,0,0)

seed(1)
# generate some random numbers

for i in range(0,100):
    sampleDate = sampleDate + timedelta(hours=1)    
    t.append(sampleDate)
    temperature.append(random())
    dewpoint.append(random()/3)
    df = df.append({'temp': random(),'dewpoint': random()/3, 'date':sampleDate},ignore_index=True)   
    #print(sampleDate)
    
print(len(t))
```

```python
import matplotlib.pyplot as plt
#plt.plot(t, temperature, 'r')
# a scatter plot comparing num_children and num_pets
df.plot(kind='line',x='date', y='temp',color='red')
plt.show()
```

```python
plt.plot(t, temperature,'r')
plt.plot(t, dewpoint,'g')
plt.xlabel('Hour  [s]')
plt.ylabel("Value [degrees]")
plt.title('Temp and dewpoint')
plt.show()
```

```python
#Using axes creates coordinate axes with figure
plt.axes([0, 0, 2, 2])
plt.plot(t, temperature,'b')
plt.xlabel('Hour  [s]')
plt.title('Temp and dewpoint')
plt.axes([0, 0, 2, 2])
plt.plot(t, dewpoint,'g')

```

```python
#Using axes creates coordinate axes with figure
plt.axes([0, 0, 2, 2])
plt.plot(t, temperature,'b')
plt.xlabel('Hour  [s]')
plt.title('Temp and dewpoint')
plt.axes([0, 0, 1, 1])
plt.plot(t, dewpoint,'g')


```
